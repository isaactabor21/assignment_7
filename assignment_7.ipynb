**What is the expected value of a single die roll?**

* The **Expected Value** or **Expectation** of a random variable $X$ is
$$
\mathbb{E}[X] = \begin{cases} \int_{x \in \text{supp}(X)} x \times f_X(x)dx, & \text{$X$ continuous}\\
\sum_{x \in \text{supp}(X)} x \times m_X(x), & \text{$X$ discrete}.
\end{cases}
$$

- A die roll, X, is discrete, so we use the discrete one where $m_x(x) = P(X=x)$.
- There are 6 possible outcomes {1,2,3,4,5,6}
- Each has probability 1/6
- $m_x(x) = 1/6$
- $
\mathbb{E}[X] = \sum_{x=1}^{6} x \times \frac{1}{6}
= \frac{1}{6}(1 + 2 + 3 + 4 + 5 + 6)
= 3.5.
$





**What is the expected value of rolling two dice and adding the results together?**

- $X_2$ = outcome of die 2

\mathbb{E}[S] = {E}[X_1 + X_2] = {E}[X_1] + {E}[X_2] $
- ${E}[X_i]$ = 3.5, so ${E}[S]$ = 7
- Expected value of rolling 2 dice and adding the results together is 7.

**Imagine that the process described in the last question continues until you fail to roll a six. What is the expected value of the process? (This can be tricky, you can simulate it to get an answer if you prefer. Hint: The answer is 4.2.)**

```python
import random

def roll_process():
    total = 0
    while True:
        roll = random.randint(1, 6)
        total += roll
        if roll != 6:
            break
    return total

# Simulate it many times
trials = 1_000_000
results = [roll_process() for _ in range(trials)]
expected_value = sum(results) / trials

print(f"Simulated expected value: {expected_value:.4f}")
```

- This is because even though you could theoretically roll sixes forever, the probability of doing so decreases exponentially:

    - Probability of 1 six in a row = 1/6

    - Probability of 2 sixes in a row = (1/6)^2 = 1/36

    - Probability of 3 sixes in a row = (1/6)^3 = 1/216

    - …and so on

- So the total expected value converges to around 4.2 still.

**Let treatment, $T$ be distributed binomial with parameter $p$. Then the contribution to the likelihood for each patient $i$, with $y_i = 0$ for no radiation therapy and $y_i=1$ for radiation therapy, is $p^{y_i}(1-p)^{1-y_i}$ Write out the likelihood.**


- $L(p) = \prod_{i=1}^n p^{y_i}(1-p)^{1-y_i}$


**Maximize the likelihood with respect to $p$. What is the MLE, $\hat{p}$?**
- log-likelihood: $\ell(p) = \sum_{i=1}^n y_i \log(p) + (1-y_i) \log(1-p)$
- Take the derivative of the log-likelihood:
- $\ell'(p) = \sum_{i=1}^n y_i \frac{1}{p} - (1-y_i) \frac{1}{1-p}$
- Set equal to 0: 
- $\sum_{i=1}^n y_i \frac{1}{\hat{p}} - (1-y_i) \frac{1}{1-\hat{p}} = 0$
- Solving for p, MLE is:
- $\hat{p} = \dfrac{1}{n} \sum_{i=1}^n y_i.$






- Bootstrap the sampling density/distribution of $\hat{p}$.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
# bootstrap the sampling density/distribution of my mu and sigma and plot their kdes
y = pd.to_numeric( y.replace({'NO':'0','YES':'1'}) , errors='coerce' ) # Convert string to numeric

n_samples = 1000
mu_boot = []
sigma_boot = []
for i in range(n_samples):
    sample = np.random.choice(y, size=len(y), replace=True)
    mu_boot.append(np.mean(sample))
    sigma_boot.append(np.std(sample))

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=mu_boot, ax=axes[0], label='mu')
sns.kdeplot(x=sigma_boot, ax=axes[1], label='sigma')
axes[0].set_title('KDE of mu')
axes[1].set_title('KDE of sigma')
plt.show()
```

- these suggest pretty reliable estimates because the kdes are pretty narrow

**Derive the density for this distribution.**

The Normal distribution has the following probability distribution function: 

$f(x) = \dfrac{1}{\sqrt{2\pi} \sigma}e^{-\frac{1}{2}\left( \frac{x-\mu}{\sigma} \right)^2}$

**Write out the likelihood.**

- The likelihood function for the Normal distribution is: 

- $L(\mu, \sigma) = \prod_{i=1}^n \dfrac{1}{\sqrt{2\pi} \sigma}e^{-\frac{1}{2}\left( \frac{x_i-\mu}{\sigma} \right)^2}.$

**Maximize the likelihood. What is the MLE?**

- Yes, maximizing the likelihood wrt $\mu$ and $\sigma$ yields their MLEs:

- First, take the log: 

- $ \ell(\mu, \sigma) = \sum_{i=1}^n \log \left( \dfrac{1}{\sqrt{2\pi}}\right) - \log(\sigma) -\frac{1}{2}\left( \frac{x_i-\mu}{\sigma} \right)^2 $

- Then, take derivatives wrt $\mu$ and $\sigma$:
$$
\dfrac{\partial \ell(\mu,\sigma)}{\partial \mu} = \sum_{i=1}^n  \frac{x_i-\mu}{\sigma^2}
$$
$$
\dfrac{\partial \ell(\mu,\sigma)}{\partial \sigma} = \sum_{i=1}^n  \frac{-1}{\sigma} +  \frac{(x_i-\mu)^2}{\sigma^3} 
$$

- Set each equal to 0 and solve: 
$$
\hat{\mu} = \frac{1}{n}\sum_{i=1}^n x_i
$$

$$
\hat{\sigma} = \sqrt{ \frac{1}{n} \sum_{i=1}^n (x_i - \hat{\mu})^2 }
$$




**Plot the density/distribution for your fitted model and compare it to the KDE/ECDF. Criticize the fit.**
